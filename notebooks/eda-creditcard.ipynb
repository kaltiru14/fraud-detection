# Credit Card Fraud Data EDA

## Objective
- Explore `creditcard.csv` dataset
- Understand feature distributions and class imbalance
- Prepare for preprocessing and feature engineering


```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler

plt.style.use("default")
sns.set_theme()
```

```python
# Load Dataset
credit = pd.read_csv("../data/raw/creditcard.csv")
credit.head()
```

```python
credit.info()
credit.describe()
```

```python
credit.isnull().sum()
```

```python
duplicates = credit.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
credit = credit.drop_duplicates()
```

```python
print("Original class distribution:")
print(credit['Class'].value_counts())

print("Class distribution (%)")
print(credit['Class'].value_counts(normalize=True) * 100)

sns.countplot(x='Class', data=credit)
plt.title("Credit Card Fraud Class Distribution (0 = Legit, 1 = Fraud)")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(credit['Amount'], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Class', y='Amount', data=credit)
plt.title("Transaction Amount by Fraud Class")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=credit[['V1','V2','V3','V4','V5']])
plt.title("Sample PCA Feature Distributions")
plt.show()
```

```python
# Scale Amount
scaler = StandardScaler()
credit['Amount_scaled'] = scaler.fit_transform(credit[['Amount']])

# Time-based feature (hour of day)
credit['hour_of_day'] = (credit['Time'] // 3600) % 24

# Preview
credit[['Time','Amount','Amount_scaled','hour_of_day']].head()
```

```python
credit.to_csv("../data/processed/creditcard_cleaned.csv", index=False)
print("Processed creditcard dataset saved to ../data/processed/creditcard_cleaned.csv")
```

## Summary of EDA and Preprocessing

**Key Findings:**
- Dataset contains no missing values, and duplicates were removed.
- Fraudulent transactions are extremely rare (~0.17%), indicating severe class imbalance.
- Fraudulent transactions tend to have higher amounts, visible in boxplots.
- PCA features (V1-V28) are already anonymized; minimal transformation required.
- Scaled `Amount` and derived `hour_of_day` for possible feature usefulness in modeling.
- Processed dataset saved for downstream feature engineering and modeling.
